UNIFORM DISTRIBUTION


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Generate a uniform distribution
uniform_data = np.random.uniform(low=0, high=10, size=100)
```

```python

# Plot the uniform distribution
plt.figure(figsize=(8, 5))
sns.histplot(uniform_data, bins=30, kde=True, color='blue')
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

BINOMIAL DISTRIBUTION

```python
# Generate a binomial distribution
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)

# Plot the binomial distribution
plt.figure(figsize=(8, 5))
sns.histplot(binomial_data, bins=30, kde=False, color='green')
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

POISSON DISTRIBUTION

```python
# Generate a Poisson distribution
poisson_data = np.random.poisson(lam=3, size=1000)

# Plot the Poisson distribution
plt.figure(figsize=(8, 5))
sns.histplot(poisson_data, bins=30, kde=False, color='orange')
plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
```

EXPONENTIAL DISTRIBUTION

```python
# Generate an exponential distribution
exponential_data = np.random.exponential(scale=1, size=1000)

# Plot the exponential distribution
plt.figure(figsize=(8, 5))
sns.histplot(exponential_data, bins=30, kde=True, color='red')
plt.title("Exponential Distribution")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
```
